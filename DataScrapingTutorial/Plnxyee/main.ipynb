```python
import os
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
os.environ['PATH'] += r"C:\A_Programs\ProjectCode\DataScrapingTutorial\Plnxyee"
driver = webdriver.Chrome()
driver.get('https://stockmarketmba.com/symbollookup.php')
```

```python
df = pd.read_csv('Nethertlands.csv').Ticker.str.split(expand = True)
df[3] = df[1] + ':' + df[0]
ticker_list = df[3].tolist()
# location_list = df[1].tolist()
sedol_list = []
```

```python
for i in range(len(ticker_list)):
    ticker = driver.find_element(By.ID, 'search')
    ticker.send_keys(ticker_list[i])
    ticker.send_keys(Keys.ENTER)
    time.sleep(0.5)

#     try:
#         location = driver.find_element(By.XPATH, '//*[@id="searchtable_filter"]/label/input')
#     except:
#         sedol_list.append('')
#     else:
#         location.send_keys(location_list[i])
#         location.send_keys(Keys.ENTER)
#         time.sleep(0.5)
    
    check = driver.find_elements(By.XPATH, '//*[@id="searchtable"]/tbody/tr[1]/td[1]')
    if (check[0].text != ticker_list[i]):
        sedol_list.append('')
   
    else: 
        sedol = driver.find_elements(By.XPATH, '//*[@id="searchtable"]/tbody/tr/td[7]')

        try:

            sedol_list.append(sedol[0].text)
        except:
            sedol_list.append('')
```

```python
df_ = pd.read_csv('Nethertlands.csv')
df_['sedol'] = sedol_list
df_.to_csv('Nethertlands_output.csv', index=False)
```
