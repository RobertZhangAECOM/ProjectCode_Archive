```python
import pandas as pd
```

```python
df = pd.read_csv('SkimsProcessing.csv')
df.head()
```

### 1. Rail

```python
x = df.columns.get_loc('Rail Access Impedances')
df_rail = df.iloc[:, x:x+5]

new_header = df_rail.iloc[0] 
df_rail = df_rail[1:]
df_rail.columns = new_header 

df_rail = df_rail.groupby('Zone').head(3).reset_index(drop = True) # the input should be sorted
df_rail.head()
```

### 2. Air

```python
x = df.columns.get_loc('Air Access Impedances')
df_air = df.iloc[:, x:x+5]

new_header = df_air.iloc[0] 
df_air = df_air[1:]
df_air.columns = new_header 

df_air = df_air.groupby('Zone').head(3).reset_index(drop = True) # the input should be sorted
df_air.head()
```

### 3. Bus

```python
x = df.columns.get_loc('Bus Access Impedances')
df_bus = df.iloc[:, x:x+5]

new_header = df_bus.iloc[0] 
df_bus = df_bus[1:]
df_bus.columns = new_header 

df_bus = df_bus.groupby('Zone').head(3).reset_index(drop = True) # the input should be sorted
df_bus.head()
```

### 4. Auto

```python
x = df.columns.get_loc('Auto Impedances')
df_auto = df.iloc[:, x:x+5]

new_header = df_auto.iloc[0] 
df_auto = df_auto[1:]
df_auto.columns = new_header 

df_auto = df_auto.reset_index(drop = True)
df_auto.head()
```

```python
df_concat = pd.concat([df_rail, df_air, df_auto, df_bus], axis=1, keys=['Rail Access Impedances', 'Air Access Impedances', 'Auto Impedances', 'Bus Access Impedances'])
df_concat.to_csv('skim_simplification_output.csv')
```
